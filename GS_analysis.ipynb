# Data cleaning and converting datatypes

```python
#Removing unwanted columns
```

```python
gs_data.drop(columns=['è®°å½æ°','Order.Priority','Row.ID','Market2'],inplace=True)
```

```python
gs_data.info()
```

```python

gs_data['Order.Date'] = pd.to_datetime(gs_data['Order.Date'], errors='coerce')
gs_data['Ship.Date'] = pd.to_datetime(gs_data['Ship.Date'], errors='coerce')

gs_data['Year'] = gs_data['Order.Date'].dt.year
gs_data['Month'] = gs_data['Order.Date'].dt.month
gs_data['Week'] = gs_data['Order.Date'].dt.isocalendar().week


gs_data['Order.Date'] = gs_data['Order.Date'].dt.strftime('%Y-%m-%d')
gs_data['Ship.Date'] = gs_data['Ship.Date'].dt.strftime('%Y-%m-%d')
```

```python
gs_data["Sales"] = gs_data["Sales"].replace({",": ""}, regex=True).astype(float)
gs_data["Profit"] = gs_data["Profit"].replace({",": ""}, regex=True).astype(float)
gs_data["Shipping.Cost"] = gs_data["Shipping.Cost"].replace({",": ""}, regex=True).astype(float)
```

```python
gs_data['Profit_margin']=(gs_data['Profit'])/(gs_data['Sales'])*100
```
